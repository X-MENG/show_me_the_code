# 找出列表中最大的数字

```python
import random
num_list = [random.randint(1, 10000) for _ in range(10)]

def FindMax(nums):
    # TODO
    max_num = nums[0]
    for i in range(1, 10):
        if nums[i] > max_num:
            max_num = nums[i]
    return max_num
    
print(num_list)
max_num = FindMax(num_list)
print('max num:', max_num)
```

# 对列表中的数字进行求和

```python
import random
num_list = [random.randint(1, 100) for _ in range(10)]

def SumList(nums):
    # TODO
    sum = 0
    for item in num_list:
        sum += item   
    return sum
    
print(num_list)
print('sum should be:', sum(num_list))
print('my sum is:', SumList(num_list))
```

# 计算列表中数字的平均值

```python
import random
num_list = [random.randint(1, 100) for _ in range(10)]

def CalcAverage(nums):
    # TODO
    ans = SumList(nums) / len(nums)
    return ans

average_num = CalcAverage(num_list)
print('average shoule be:', sum(num_list) / len(num_list))
print('average :', average_num)
```

# 将任意范围内的偶数保存到列表中

```python
def Evens2List(first, last):
    ret = []
    # TODO
    for i in range(first, last + 1):
        if i % 2 == 0:
            ret.append(i)
    return ret

lst = Evens2List(1, 100)
print(lst)
```

# 实现一个函数，这个函数的功能是：传入一个数字num, 计算这个数字的约数之和，这里的约数不包含自身

```python
def CalcFactorSum(num):
    #TOD0
    sum = 0
    for i in range(1, num):
        if num % i == 0:
            sum += i
    return sum

# TEST
sum = CalcFactorSum(123456789)
print('sum =', sum)
```
