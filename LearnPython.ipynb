# 找出列表中最大的数字

```python
import random
num_list = [random.randint(1, 10000) for _ in range(10)]

def FindMax(nums):
    # TODO
    max_num = nums[0]
    for i in range(1, 10):
        if nums[i] > max_num:
            max_num = nums[i]
    return max_num
    
print(num_list)
max_num = FindMax(num_list)
print('max num:', max_num)
```

# 对列表中的数字进行求和

```python
import random
num_list = [random.randint(1, 100) for _ in range(10)]

def SumList(nums):
    # TODO
    sum = 0
    for item in num_list:
        sum += item   
    return sum
    
print(num_list)
print('sum should be:', sum(num_list))
print('my sum is:', SumList(num_list))
```

# 计算列表中数字的平均值

```python
import random
num_list = [random.randint(1, 100) for _ in range(10)]

def CalcAverage(nums):
    # TODO
    ans = SumList(nums) / len(nums)
    return ans

average_num = CalcAverage(num_list)
print('average shoule be:', sum(num_list) / len(num_list))
print('average :', average_num)
```

# 输出1-100之间所有的偶数

```python
def output_even_jack():
    for i in range(1, 101):
        if i % 2 == 0:
            print(i, end = ' ')
output_even_jack()

print('\n')

def output_even_dad():
    a = 2
    while(a <= 100):
        print(a, end = ' ')
        a += 2

output_even_dad()
```
